```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

energy=[]
pcone = []
pctwo = []
with open('pot-energy.csv', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for rows in plotting:
        energy.append(float(rows[0]))

#Remove the first row with labels in pcaINFO-resFLARE.csv file!
filename="pcaINFO-resFLARE.csv"

df= pd.read_csv(filename, names=['Min. bond length (Å) (H–C)','Max. bond length (Å) (H–C)','Mean bond length (Å) (H–C)','Pop. std. dev. (H–C)','Std. error (H–C)','Min. bond length (Å) (S–C)','Max. bond length (Å) (S–C)','Mean bond length (Å) (S–C)','Pop. std. dev. (S–C)','Std. error (S–C)','Min. bond length (Å) (S–Au)','Max. bond length (Å) (S–Au)','Mean bond length (Å) (S–Au)','Pop. std. dev. (S–Au)','Std. error (S–Au)','Min. bond length (Å) (Au–Au)','Max. bond length (Å) (Au–Au)','Mean bond length (Å) (Au–Au)','Pop. std. dev. (Au–Au)','Std. error (Au–Au)','POC for 0','POC for 1','POC for 2','POC for 3','POC for 4','POC for 5','POC for 6','POC for 7','POC for 8','POC for 9','POC for 10','POC for 11','POC for 12','POC for 13','POC for 14','POC for 15','0°-20°','20°-40°','40°-60°','60°-80°','80°-100°','100°-120°','120°-140°','140°-160°','160°-180°','target'])

features = ['Min. bond length (Å) (H–C)','Max. bond length (Å) (H–C)','Mean bond length (Å) (H–C)','Pop. std. dev. (H–C)','Std. error (H–C)','Min. bond length (Å) (S–C)','Max. bond length (Å) (S–C)','Mean bond length (Å) (S–C)','Pop. std. dev. (S–C)','Std. error (S–C)','Min. bond length (Å) (S–Au)','Max. bond length (Å) (S–Au)','Mean bond length (Å) (S–Au)','Pop. std. dev. (S–Au)','Std. error (S–Au)','Min. bond length (Å) (Au–Au)','Max. bond length (Å) (Au–Au)','Mean bond length (Å) (Au–Au)','Pop. std. dev. (Au–Au)','Std. error (Au–Au)','POC for 0','POC for 1','POC for 2','POC for 3','POC for 4','POC for 5','POC for 6','POC for 7','POC for 8','POC for 9','POC for 10','POC for 11','POC for 12','POC for 13','POC for 14','POC for 15','0°-20°','20°-40°','40°-60°','60°-80°','80°-100°','100°-120°','120°-140°','140°-160°','160°-180°']

x = df.loc[:, features].values
y = df.loc[:,['target']].values
x = StandardScaler().fit_transform(x)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)

#the range can be vary depending on the number of datapoints you have
#for the example we provided use range(0, 5)
for i in range(0, 100):
    pcone.append(principalComponents[i,0])
    pctwo.append(principalComponents[i,1])

principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])
finalDf = pd.concat([principalDf, df[['target']]], axis = 1)

#for 3D plots
fig = plt.figure(figsize = (12.0,10.0))
ax = plt.axes(projection='3d')
ax.set_xlabel('$PC_{1}$', fontsize=17, labelpad=15.0)
ax.set_ylabel('$PC_{2}$', fontsize=17, labelpad=15.0)
ax.set_zlabel('$E_{pot}$ (eV)', fontsize=17, labelpad=35.0)
ax.set_xlim3d(-11, 16)
ax.set_ylim3d(-6, 10)
#ax.set_zlim3d(-100, 100)
plt.yticks(fontsize=17)
plt.xticks(fontsize=17)

#changing the tick label padding
ax.tick_params(axis='z', which='major', pad=15)

font = {'size': 17}
ax.tick_params('z', labelsize=font['size'])

#change the view angle!
ax.view_init(15,-52.5)

fig.set_facecolor('white')
ax.set_facecolor('white')

# plt.title('3D Principal component analysis')
plt.ticklabel_format(style='sci')
ax.scatter(pcone, pctwo, energy, alpha=1.0, s=32.5, color="#b30000")

ax.plot(pcone, pctwo, 'o', markersize=4.0, alpha=0.8, color='#5ad45a', markeredgecolor='none', zdir='z', zs=-336.2)
ax.plot(pcone, energy, 'o', markersize=4.0, alpha=0.8, color="#1a53ff", markeredgecolor='none', zdir='y', zs=10.0)
ax.plot(pctwo, energy, 'o', markersize=4.0, alpha=0.8, color="#ffa300", markeredgecolor='none', zdir='x', zs=-10.0)

plt.savefig('updated-principal-component-analysis-3D.png',dpi=1080)#,transparent=True)


#following section will save a CSV file with PC1, PC2, and the energy information
# create a dictionary with the three lists
dict = {'PCAone': pcone, 'PCAtwo': pctwo, 'Energy': energy}
# create a Pandas DataFrame from the dictionary
df = pd.DataFrame(dict)
# write the DataFrame to a CSV file
df.to_csv('PCA-info.csv')
```
